```python
#Se crea un data frame de la SUMA de todos los valores de las columnas del barrio y se almacena en una variable
datos_barrio = inmuebles.groupby('Barrio').sum()
datos_barrio
```

```python
##a ese data frame se le crea una columna que va a ser el promedio de la suma de Precio por millon / Area y ese es el promedio de precio-m2 por Barrio
datos_barrio['Precio_M2_Barrio'] = datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

```python
#Se crea un diccionario en base al data frame con los valores de la columna creada "Precio_M2_Barrio"
m2_barrio=dict(datos_barrio['Precio_M2_Barrio']) 
```

```python
#Se crea una nueva columna en el data set original para agregar la sustitución de los valores de la columna del data frame de promedio m2 por el antiguo promedio de promedio precio m2 por tipo
inmuebles['Precio_M2']=inmuebles['Barrio']
inmuebles['Precio_M2']=inmuebles['Barrio'].map(m2_barrio)
inmuebles.head(3)
```

```python
#Para conseguir el top 10 de algo
top_barrios = inmuebles['Barrio'].value_counts()[:10].index #Devuelve solo la lista de indices del top 10 con mas inmuebles
top_barrios
```

```python
#datos_barrio #El data frame datos barrio lo vamos a resetear para que Barrio sea una columna y no un indice
datos_barrio.reset_index(inplace=True) #Con reset index y el inplace permite reasignar directamente
datos_barrio
```

```python
#La funcion query sirve para consultar un dato dentro de una variable, en este caso los barrios que esten en TOP_BARRIOS creada anteriormente
datos_barrio.query('Barrio in @top_barrios')
```

```python
#Creamos el lienzo para la grafica, se utilizara un box plot, utilizamos el query como data
plt.figure(figsize=(10,8))
ax =sns.barplot(x="Barrio",y="Precio_M2_Barrio", data=datos_barrio.query('Barrio in @top_barrios'), palette="rocket")
ax.tick_params(axis='x', rotation=45)
```
